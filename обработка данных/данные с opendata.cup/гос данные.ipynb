```python
import pandas as pd
import os
```

```python
data = []
for i in os.listdir():
    if i.endswith(".csv"):
        data.append(pd.read_csv(i))
        data[-1]["type"] = i[:-4]
```

```python
data[0] = data[0].rename({"Наименование": "название"}, axis=1)
data[1] = data[1].rename({"Объекты": "название"}, axis=1)
data[2] = data[2].rename({"Наименование": "название"}, axis=1)
data[3] = data[3].rename({"Наименование": "название"}, axis=1)
```

```python
columns = ["название", "адрес", "Долгота", "Широта", "type"]
```

```python
df = pd.concat([i[columns] for i in data])
```

```python
df["название"] = df["название"].str.strip()
df["адрес"] = df["адрес"].str.strip()
```
